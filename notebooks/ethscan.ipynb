```python
import os

import pandas as pd
import requests

import datetime
```

```python
ETHERSCAN_API_KEY = os.getenv("ETHERSCAN_API_KEY")
```

```python
# get current time - 1 day in unix timestamp
timestamp = round((datetime.datetime.now() - datetime.timedelta(days=1)).timestamp())

url = f"https://api.etherscan.io/api?module=block&action=getblocknobytime&timestamp={timestamp}&closest=before&apikey={ETHERSCAN_API_KEY}"

response = requests.get(url)
response.raise_for_status()
response.json()

# get block number
block_number = response.json()["result"]
block_number
```

0xA88800CD213dA5Ae406ce248380802BD53b47647 is the settlement contract

```python
address = "0xA88800CD213dA5Ae406ce248380802BD53b47647"

url = f"https://api.etherscan.io/api?module=account&action=txlist&address={address}&startblock={block_number}&endblock=99999999&page=1&offset=10000&sort=asc&apikey={ETHERSCAN_API_KEY}"

response = requests.get(url)
response.json()
```

```python
transactions = response.json()["result"]
transactions[0].keys()
```

```python
# show all dataframe columns 
pd.set_option('display.max_columns', None)
```

```python

df = pd.DataFrame(transactions)
df.head()
```

```python
df.dtypes
```

```python
# convert appropriate columns to numeric
df = (
    df
    .assign(
        blockNumber=pd.to_numeric(df["blockNumber"]),
        timeStamp=pd.to_numeric(df["timeStamp"]),
        gas=pd.to_numeric(df["gas"]),
        gasPrice=pd.to_numeric(df["gasPrice"]), # in wei
        gasUsed=pd.to_numeric(df["gasUsed"]),
        cumulativeGasUsed=pd.to_numeric(df["cumulativeGasUsed"]),
        nonce=pd.to_numeric(df["nonce"]),
        transactionIndex=pd.to_numeric(df["transactionIndex"]),
        value=pd.to_numeric(df["value"]),

        # isError, and in this case, even from and to can be considered as categorical
        isError=df["isError"].astype("category"),
        to=df["to"].astype("category"),
        **{"from": df["from"].astype("category")},

        # convert timestamp to datetime
        datetime=pd.to_datetime(df["timeStamp"], unit="s")
    )
)

df.dtypes
```

```python
(
    df
    .query("isError == '0' and methodId == '0x0965d04b'")
    .assign(transactionCost=lambda x: x["gasUsed"] * x["gasPrice"] / 1e18)
    .groupby("from")
    .agg(
        total_gas_used=("gasUsed", "sum"),
        avg_gas_used=("gasUsed", "mean"),
        avg_gas_price=("gasPrice", "mean"),
        total_transaction_cost=("transactionCost", "sum"),
        avg_transaction_cost=("transactionCost", "mean"),
    )
)
```

```python
transaction_hash = "0x604083aab3c7edd1602ecd00a6cc260e926f46dd04e5e6a28b2112f679755d08"

url = (f"""https://api.etherscan.io/api
   ?module=proxy
   &action=eth_getTransactionReceipt
   &txhash={transaction_hash}
   &apikey={ETHERSCAN_API_KEY}"""
   .replace("\n", "").replace(" ", "")
)


response = requests.get(url)
response.json()
```

```python
receipt = response.json()["result"]
logs = receipt["logs"]
logs_df = pd.DataFrame([log for log in logs if len(log["topics"]) == 3])
logs_df[['topic_hash', 'from_address', 'to_address']] = pd.DataFrame(logs_df['topics'].tolist(), index=logs_df.index)
logs_df = logs_df.drop(columns=['topics'])

logs_df
```

```python
def unpad_address(address):
    # remove the 0x prefix
    address = address[2:]
    # remove the leading 0s
    address = address.lstrip("0")
    # add the 0x prefix back
    address = "0x" + address
    return address
```

```python
(
    logs_df
    .assign(
        from_address=lambda x: x["from_address"].map(unpad_address),
        to_address=lambda x: x["to_address"].map(unpad_address),
    )
    # convert data from hex to int
    .assign(data=lambda x: x["data"].map(lambda x: int(x, 16)))
    
    # drop irrelevant columns
    .drop(columns=["blockHash", "blockNumber", "transactionHash", "removed", "logIndex"])
)
```
